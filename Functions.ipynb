 1. What is the difference between a function and a method in Python?

- In Python, a function is a reusable unit of code that performs a specified task and is declared with the def keyword. Functions can exist independently and are not associated with any object. You can call a function by name, and it may or may not accept parameters or return values. A method, on the other hand, is a function that is associated with an object and called using dot notation. Methods are defined within classes and automatically use the instance (typically self) as their first parameter. The main distinction is that functions exist independently, whereas methods are functions that belong to objects (class instances). For example, len() is a function, whereas "hello".upper() is a method called on a string object.

2. Explain the concept of function arguments and parameters in Python.

- Parameters and arguments are terminology used in Python to describe functions, and while they are related, they mean different things. Parameters are the variables listed in a function's definition that serve as placeholders for the values that the function anticipates receiving. Arguments, on the other hand, are the actual values that the function receives when it is called. For example, in the function definition def greet(name), name is a parameter. When you call the function with greet("Alice"), the parameter is "Alice". Python accepts a variety of parameters, including positional, keyword, default, and variable-length arguments, providing flexibility in how functions are constructed and called. Understanding the distinction between parameters and arguments is essential for writing functions that are both reusable and adaptable.

3. What are the different ways to define and call a function in Python?

Another method for defining a function is to utilise lambda functions, which are anonymous and intended for simple, quick tasks. These different calling methods allow developers to write flexible and reusable code that can handle a variety of input scenarios.

4.  What is the purpose of the `return` statement in a Python function?

- The return statement in a Python function is used to send the result of the function’s execution back to the caller. When a function reaches a return statement, it exits immediately, and the value specified is passed back as the output of the function. This allows functions to produce results that can be stored in variables, used in expressions, or passed to other functions. If no return statement is used, the function returns None by default. 5.  What are iterators in Python and how do they differ from iterables?

- In Python, an iterable is any object that can be looped over, such as lists, tuples, strings, and dictionaries. These objects implement the __iter__() method, which returns an iterator. An iterator is an object that keeps track of its position during iteration and implements two methods: __iter__() and __next__(). The __next__() method returns the next item in the sequence and raises a StopIteration exception when there are no more items. The key difference is that iterables can be looped over multiple times (by creating a new iterator each time), while iterators are consumed once and cannot be reused without creating a new one. In simple terms, iterables are data structures you can iterate through, while iterators are the tools that actually perform the iteration.

6.  Explain the concept of generators in Python and how they are defined.

- In Python, generators are a sort of iterable that allows you to iterate over data one item at a time without storing the complete sequence in memory. They are defined as functions that use the yield keyword instead of return. When the generator function is called with next(), execution continues from where it left off, preserving its internal state. This makes generators extremely memory efficient, particularly when working with big datasets or infinite sequences. Unlike conventional functions, which return a single value and depart, a generator can produce several values one at a time, waiting between each yield until the next value is requested. For example, a generator that yields numbers from 1 to 3 would be defined as defcount (): yield 1; yield 2; yield 3. Generators are commonly used in loops and are ideal for lazy evaluation and streaming data processing.

7. What are the advantages of using generators over regular functions?

- Generators have various advantages over normal functions, especially when working with huge datasets or streams of data. The main advantage is memory efficiency, as generators produce values one at a time using the yield keyword, rather to computing and storing a full series in memory at once. This makes them useful for jobs such as reading big files and creating infinite sequences. Another advantage is lazy evaluation, which generates values only when they are required, so improving performance and reducing overhead. Generators also make code cleaner and more understandable when working with sequences, particularly in loops. Generators can also simplify complex iteration logic by maintaining their state between yields, which eliminates the need for additional variables or data structures. Overall, generators help write more efficient and elegant code in scenarios where full data storage is necessary.

8. What is a lambda function in Python and when is it typically used?

- A lambda function in Python is a tiny, anonymous function written with the lambda keyword rather than def. It accepts any number of inputs but can only contain one expression, which is evaluated and returned. Lambda functions are commonly used for brief, simple actions that are only required temporarily, particularly as parameters to higher-order functions such as map(), filter(), and sorted(). For instance, lambda x: x * 2 defines a function that doubles its argument. Lambda functions are useful in cases where specifying a whole function with def would be unneeded or make the code less clear. Regular functions are preferable for more sophisticated tasks or code reuse.

9.  Explain the purpose and usage of the `map()` function in Python.

- The map() function in Python applies a provided function to each item in an iterable (such as a list or tuple) and produces a map object, which is an iterator that contains the results. Its principal role is to change data by applying a function to all items of a collection without requiring an explicit loop. The syntax is map(function, iterable), which applies the function to each element of the iterable. For example, map(lambda x: x * 2, [1, 2, 3]) would result in an iterator containing [2, 4, 6]. Because map() outputs a lazy iterator, it is frequently converted to a list using list() to display the results. This function is especially useful for writing clean, concise, and functional-style code when transforming or processing collections.

10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

- Python's built-in functions map(), reduce(), and filter() facilitate functional programming by processing iterables in various ways. The map() function applies a provided function to each item in an iterable and returns an iterator of the results, so altering each element. In contrast, filter() uses a function to produce a boolean value for each element, and only elements that evaluate to True are included in the output, allowing you to selectively filter items from the iterable. Reduce(), which is included in the functools module, accepts a function and a series and applies the function cumulatively to reduce the sequence to a single value—for example, summing a list of numbers. While map () transforms, filter () selects, and reduce () combines-all three provide concise alternatives to traditional loops and can lead to more readable and function-style code when used appopriately.

1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

- Here is a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list:

```python
def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum
```

4.  Write a Python function that checks if a given number is prime or not from 1 to 200

- Here is a Python function that checks whether a given number (from 1 to 200) is a prime number:

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

- Here’s how you can create an iterator class in Python to generate the Fibonacci sequence up to a specified number of terms:

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a
```

6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.

- Here’s a generator function in Python that yields powers of 2 up to a given exponent:

```python
def powers_of_two(max_exponent):
    for i in range(max_exponent + 1):
        yield 2 ** i
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

- Here’s a generator function in Python that reads a file line by line and yields each line as a string:

```python
def read_file_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')
```

8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

- You can use a lambda function with Python’s sorted() function to sort a list of tuples based on the second element of each tuple like this:

```python
# Sample list of tuples
tuple_list = [(1, 3), (4, 1), (2, 2), (5, 0)]

# Sort based on the second element using lambda
sorted_list = sorted(tuple_list, key=lambda x: x[1])

print("Sorted list:", sorted_list)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

- Here’s a Python program that uses the map() function to convert a list of temperatures from Celsius to Fahrenheit:

```python
# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Conversion formula: (C × 9/5) + 32 = F
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

# Print the results
print("Temperatures in Fahrenheit:", fahrenheit_temps)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

- Here’s a Python program that uses the filter() function to remove all the vowels from a given string:

```python
def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_str))

# Example usage
text = "Hello World"
result = remove_vowels(text)
print("String without vowels:", result)
```
